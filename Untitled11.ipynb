Glyphs in Bokeh are visual markers used to represent data points in a plot. 

They can be used to create scatter plots, line plots, bar charts, and many other types of plots

```python
from bokeh.plotting import figure, output_file, show


x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


output_file("plot.html")


p = figure(title="Example Plot", x_axis_label="X-axis", y_axis_label="Y-axis")
p.circle(x, y, size=10, color="red", alpha=0.5)


show(p)

```
