# Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.

Simple linear regression is a technique of regression where  prediction of a label is made with the help of a single feature and single target.
the regression line obtained here is a single line,so the intercept and the coef obtained here is only one.
The general formula for representation of simple linear regression is :y=c+mx

Multiple linear regression is a technique of regression where predictio of lable is made with the help of more than one feature.
According to dimension the regression is obtained it may be plane for 2 features and the dimension of the regression also changes 
according to dimension of features and labels. Here, we get multiple intercepts and coeff .
The formula to represent multiple linear regression is : 
    y=c+mx1+mx2+mx3+........
    




Example of simple linear regression is :
    predicting  the age of  a person according to their weights.
    
Example of multiple linear regression is:

# Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?

The assumptions of linear regression are:

    1) Linear relationship between input and output:For using linear regression the data in the inputs must be linear with the 
    
        datapoints of the labels.It is applicable for single as well as multiple regression.In multiple regression also, every
        
        feature   must show linear relationship with the labels.
        
    2) No Multicollinearity:
    If you have multiple features like x1,x2,x3,x4,x5 then there should not be any correlation between the input features.
    
        The input features must be independent of themselves if one change in one creates a change in other then it not valid.
        
    3) Normal Residulas:
    
    If you plot a kde of a residual then the graph should show a normal distribution.
    
        It is because in the errors(residuals) the variance must be 1.
        
    4)Homoscedasticity:
   
   If you plot the scatter plot of residual(y-axis) with y_predictions(x-axis) then the scatterplot must be uniformly
   
        scatterred.
        
    5) No Autocorrelation of residuals:
    
    If you plot the residuals then the plot should be zig-zag and not form any pattern.
    

# Checking of assumptions

#### Linear relationships between inputs and output

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/campusx-official/linear-regression-assumptions/main/data.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2,3,1)
plt.scatter(df['feature1'],df['target'])
plt.subplot(2,3,2)
plt.scatter(df['feature2'],df['target'])
plt.subplot(2,3,3)
plt.scatter(df['feature3'],df['target'])
```

#### No multicollinarity

```python
sns.heatmap(df.corr(),annot=True)
```

#### Normal Residuals

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=1)
```

```python
regressor=LinearRegression()
model=regressor.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
residuals=y_test-y_pred
```

```python
sns.displot(residuals,kind='kde')
```

#### Homoscedasticity


```python
sns.scatterplot(y_pred,residuals)
```

#### No autocorrelation of residuals

```python
plt.plot(residuals)
```

# Q7. Describe the polynomial regression model. How is it different from linear regression?

Polynomial regression model is defined as the regression model in which the output feature is related with the input feature 
as:
    y=c+c1x^2+c2x^3
    
In linear regression the output is releated as :
    
    y=c1+c2X
    
In polynomial regression a curve joins the points to fit the data where as in linear regression a straight line joins the 
points.
