```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y,color='red',alpha=.8)
plt.grid()
plt.figure(figsize=(5,3))
```

Scatter plot is a graph plot in statistics that represents how the datas are correleated.

3

SubPlot() is used to plot the differnt graphs into a single one so that it will be easier for us to compare them.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color='red')
plt.grid()
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title("Profit Comparison")
```

Bar Plot is a graph plot in Statistics which is used to represent data in the form of Bar, The end points of height of bars represents the maximum range of the x-axis variable.

```python
plt.barh(company,profit,color='red')
plt.grid()
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title("Profit Comparison")
```

5

BoxPlot is a plot in statistics that plots a information as a box and the area or width of the box is the range of the values of 1st Quartile, Median and 3rd quartile 

```python
box1 = np.random.normal(100, 10, 200)
plt.subplot(2,2,1)
plt.boxplot(box1,patch_artist=True)
box2 = np.random.normal(90, 20, 200)
plt.subplot(2,2,2)

plt.boxplot(box2)
```
