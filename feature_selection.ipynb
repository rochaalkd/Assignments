# 1.What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its application.

Min-Max method is the method of Normalization in feature scaling. It is mainly done in deep learning image classification types 

 algorithms. It is method of converting the datapoints in a data sets between 0 and 1. 
    

# Example

```python
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
#Applying MinMaxScaler in fare column
scaler=MinMaxScaler()
```

```python
df['scaled_data']=scaler.fit_transform(df[['fare']])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
plt.hist(df['fare'])
plt.subplot(2,2,2)
plt.hist(df['scaled_data'])
```

# What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an example to illustrate its application.

Principle Component Analysis is the technique of reducing the dimensions of the datas in to only nth dimensions. 


For dimensionality reduction it creates a plane passing through the feature points and 

projects that feature points into that
 
Plane of nth dimension.
