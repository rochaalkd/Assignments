```python
import seaborn as sns
```

1

Plots that can be plotted using seaborn are: 


i)Scatter Plot
-Used to plot a diagram between two variables which shows how much vairable is affected by another.


ii)Histogram
-Used to plot mainly categorical data where the x-axis represents the category and y-axis the frequency of that data


iii)Relationship Plots
-Used to plot datas that shows the relationships of two or more columns where the axis are of different columns


iv)PairPlots
-Used to plot graphs between the integral columns all at once


v)Joint Plots
-Used to plot scatter plot  and histogram in the same chart

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.relplot(data=fmri,x='timepoint',y='signal',hue='event',style='region',kind='line')
```
